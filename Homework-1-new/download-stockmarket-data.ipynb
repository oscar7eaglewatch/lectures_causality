```python
import yfinance as yf
import pandas as pd

# Define the ticker symbols in a list
tickers = ["AAPL", "NVDA", "INTC", "QCOM", "CL=F", "NG=F", "USD"]

# Define the start and end dates
start_date = "2021-12-31"
end_date = "2022-06-17"

# Fetch data for all tickers using yfinance
data = yf.download(tickers, start=start_date, end=end_date, group_by="ticker")

# Create an empty DataFrame with the index from one of the fetched data sets (all should have the same index)
combined_df = pd.DataFrame(index=data[tickers[0]].index)

# Extract the 'Close' column for each ticker, rename it to the ticker's name, and add to the combined DataFrame
for ticker in tickers:
    combined_df[ticker] = data[ticker]['Close']

# Save the combined DataFrame to a CSV file
combined_df.to_csv("combined_data.csv")

print(f"Combined 'Close' data for {', '.join(tickers)} between {start_date} and {end_date} saved to 'combined_data.csv'.")

```
